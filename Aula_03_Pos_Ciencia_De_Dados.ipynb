```python
from numpy.random import seed
from numpy.random import normal
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def gerar_valores(dias):
  seed(1)
  dados = normal(loc=100, scale=10, size=dias)
  return dados
```

```python
def gerar_dados(posicao_grafico):
  ndias = 365
  dias = np.arange(365)
  y = gerar_valores(ndias)
  linha  = 2
  coluna = 3
  plt.subplot(linha, coluna, posicao_grafico)
  plt.plot(dias,y)
  return y
```

```python
def analise_exploratoria(dados, investimento):
  media = dados.mean()
  desvio_padrao=dados.std()
  minimo = dados.min()
  maximo = dados.max()
  print("-----------------------------------------------")
  print("Análise do Investimento:{}".format(investimento))
  print("-----------------------------------------------")
  print("media:{}".format(media))
  print("desvio padrão:{}".format(desvio_padrao))
  print("mínimo:{}".format(minimo))
  print("máximo:{}".format(maximo))
```

```python
import matplotlib.pyplot as plt
import numpy as np

retorno_1 = gerar_dados(1)
retorno_2 = gerar_dados(2)
retorno_3 = gerar_dados(3)
retorno_4 = gerar_dados(4)
retorno_5 = gerar_dados(5)
retorno_6 = gerar_dados(6)
plt.show()
analise_exploratoria(retorno_1,1)
analise_exploratoria(retorno_2,2)
analise_exploratoria(retorno_3,3)
analise_exploratoria(retorno_4,4)
analise_exploratoria(retorno_5,5)
analise_exploratoria(retorno_6,6)
```

**Implementar uma função para resolver o problema**

```python
def funcao_somatorio_k_k2(n):
  soma = 0
  for i in range(n):
    k = i +1
    soma = soma + 3*k + 2*k**2
  return soma
```

```python
resultado = funcao_somatorio_k_k2(10)
print('resultado: {}'.format(resultado))
```

**Dada função F definida como:**

> F(1) = 1

> F(N) = F(N/2) + 1,  para N ≥ 2.


**Suponha N= 1, 2, 4, 8, 16, 32, 64, ...**

**Implementar uma função para resolver o problema.** 


```python
def f(n):
  if(n==1):
    return 1
  else:
    return f(n/2)+1
```

```python
resultado = f(64)
print('resultado: {}'.format(resultado))
```

**Obtenha f(n) para cada elemento da lista = [2, 16, 64, 256].**

```python
lista = [2, 16, 64, 256]
for a in lista:
  print('resultado: {}'.format(f(a)))
```

```python
lista = [2, 16, 64, 256]
n = len(lista)
for i in range(n):
  print('resultado: {}'.format(f(lista[i])))
```

**Obter uma lista para dos resultados de f(n) para a lista = [2, 16, 64, 256].**

```python
resultado = []
lista = [2, 16, 64, 256]
for elem in lista:
  resultado.append(f(elem))
```

```python
print('resultado: {}'.format(resultado))
```
